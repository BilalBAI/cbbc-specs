```python
import matplotlib.pyplot as plt

def plot(top_left,top_right,bottom_left,bottom_right,x_ax,y_ax="Off-Chain to On-Chain"):
    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(8, 8))

    # Fill each quadrant with color
    ax.fill_between([-1, 0], -1, 0, color='#ff9999', alpha=0.5)  # Top left - Light Red
    ax.fill_between([0, 1], 0, 1, color='#99ccff', alpha=0.5)   # Top right - Light Blue
    ax.fill_between([-1, 0], 0, 1, color='#99ff99', alpha=0.5) # Bottom left - Light Green
    ax.fill_between([0, 1], -1, 0, color='#ffcc99', alpha=0.5)  # Bottom right - Light Orange

    # Draw the quadrant lines
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)

    # Add the labels for each quadrant
    ax.text(-0.75, 0.75, top_left, ha='left', va='center', fontsize=12, weight='bold')
    ax.text(0.75, 0.75, top_right, ha='right', va='center', fontsize=12, weight='bold')
    ax.text(-0.75, -0.75, bottom_left, ha='left', va='center', fontsize=12, weight='bold')
    ax.text(0.75, -0.75, bottom_right, ha='right', va='center', fontsize=12, weight='bold')

    # Set axis limits and labels
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)

    ax.set_xlabel(x_ax)
    ax.set_ylabel(y_ax)


    # Remove tick marks
    ax.set_xticks([])
    ax.set_yticks([])


    # Show the chart
    # plt.title("Customization vs Standardization and Off-Chain vs On-Chain", fontsize=14, weight='bold')
    plt.show()
```

```python
top_left="Customized\n& On-Chain\n\n(e.g. Aveo OTC)"
top_right="Standard\n& On-Chain\n\n(e.g. Ribbon/Aevo)"
bottom_left="Customized\n& Off-Chain\n\n(e.g. OTC Options)"
bottom_right="Standard\n& Off-Chain\n\n(e.g. Deribit Options)"
x_ax="Customized and Standard"
plot(top_left,top_right,bottom_left,bottom_right,x_ax,y_ax="Off-Chain to On-Chain")
```

```python
top_left="Exotic\n& On-Chain\n\n(On-Chain CBBC)"
top_right="Vanilla\n& On-Chain\n\n(e.g. Aevo)"
bottom_left="Exotic\n& Off-Chain\n\n(e.g. CBBC traded in HK)"
bottom_right="Vanilla\n& Off-Chain\n\n(e.g. Deribit Options)"
x_ax="Exotic and Vanilla Options"
plot(top_left,top_right,bottom_left,bottom_right,x_ax,y_ax="Off-Chain to On-Chain")
```

```python
top_left="Traditional Asset\n& On-Chain\n\n"
top_right="Cryptos\n& On-Chain\n\n"
bottom_left="Traditional Asset\n& Off-Chain\n\n(e.g. CME)"
bottom_right="Cryptos\n& Off-Chain\n\n(e.g. Deribit)"
x_ax="Underlying of the Options: Traditional Asset and Crypto Asset"
plot(top_left,top_right,bottom_left,bottom_right,x_ax,y_ax="Off-Chain to On-Chain")
```
